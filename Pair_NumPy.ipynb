Crea un array vacío con la misma forma y dimensiones que el array creado anteriormente. 
Vamos a reemplazar los valores de la matriz vacía que nos acabamos de crear pero basándonos en los valores del primer array bajo las siguientes condiciones:

- Si el valor del array es igual que la media, se reemplaza por "A".
- Si el valor del array es mayor que la media, se reemplaza por "B".
- Si el valor del array es menor que la media, se reemplaza por "C".
- Si el valor es igual que el máximo, se reemplaza por "D".
- Si el valor es igual que el mínimo, se reemplaza por "E".

```python
a5 = np.empty((2,3,5))
a5
```

```python
media_redondeada = round(np.mean(a5), 0)  # Usamos un valor entero para simplificar la comparación
valor_max = np.max(a5)
valor_min = np.min(a5)
```

```python
# Array inicial para las letras (debe ser de tipo string)
array_letras = a5.astype(str)
```

```python
# 1. Asignar el valor MÁXIMO ("D") y el resto permanece igual.
# Si la condición es Falsa, mantiene el array_letras original.
array_letras = np.where(a5 == valor_max, "D", array_letras)

# 2. Asignar el valor MÍNIMO ("E")
# Si la condición es Falsa, mantiene el resultado del paso 1.
array_letras = np.where(a5 == valor_min, "E", array_letras)

# 3. Asignar la MEDIA ("A")
# Si la condición es Falsa, mantiene el resultado del paso 2.
array_letras = np.where(a5 == media_redondeada, "A", array_letras)

# 4. Asignar Mayor que la media ("B")
# Si la condición es Falsa, mantiene el resultado del paso 3.
array_letras = np.where(a5 > media_redondeada, "B", array_letras)

# 5. Asignar Menor que la media ("C")
# Ya que todas las condiciones anteriores se han verificado, la única opción restante
# para los valores que aún no han sido sustituidos es que sean menores que la media,
# pero lo haremos explícitamente para mayor claridad.
array_letras = np.where(a5 < media_redondeada, "C", array_letras)
```

```python
array_letras
```
