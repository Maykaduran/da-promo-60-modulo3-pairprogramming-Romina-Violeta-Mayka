```python
# ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.
```

```python
df.info()
```

```python
df.groupby('continent')['country'].count().reset_index()
```

```python
sns.countplot(data=df, y='continent')
```

```python
plt.pie(data = df.groupby('continent')['country'].count().reset_index(), x= 'country', labels='continent', autopct='%1.1f%%')
plt.title('Porcentaje de países por continente')
```

```python
plt.pie(df['continent'].value_counts(), labels=df['continent'].value_counts().index, autopct='%1.1f%%');
```

```python
# ¿Cuáles son los 10 países con la mayor área forestal?
```

```python
df.groupby('country')['forested_area'].value_counts().sort_values(ascending=False).head(10)
```

```python
df.groupby('country')['forested_area'].sum().sort_values(ascending=False).head(10)
```

```python
dfarea = df.groupby('country')['forested_area'].sum().sort_values(ascending=False).head(10)
```

```python
sns.barplot(x ='forested_area', y = dfarea.index, data = dfarea.reset_index(), color = 'green')
plt.title('Top 10 países con mayor área forestal')
plt.xlabel('Área forestal (en km²)')
```

En los scatterplot los puntos deben tener un tamaño de 2.
